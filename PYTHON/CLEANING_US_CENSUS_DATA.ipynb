```python
# .iat() works with interger-based positioning.
# it is an accessor for retrieving a single value from a DataFrame by its interger location using row index and column index)
# it is used for faster access to a single value and works best with interger based positioning. 
# The syntax is : Dataframe.iat[row_index,column_index]
```
