```python
# importing the cart file
cart = pd.read_csv("C:\\Users\\Lenovo\\OneDrive\\Documents\\Desktop\\CODECADEMY\\PYTHON PROJECTS INSTRUCTIONS\\Page_Visits_Funnel_Project\\Page_Visits_Funnel_Project\\cart.csv")
cart.head()
```

```python
# Doing a merge on cart and visits 
cart_visits = visits.merge(cart, how = "left")
cart_visits.head()
```

```python
# Finding the length of cart visits
total_cart_visits = len(cart_visits)
total_cart_visits 
```

```python
# Finding the null timestamps for cart_time
cart_visits.isna().sum()
```

```python
# Dropping all the missing values
cart_visits.dropna()
```

```python
cart_visits.info()
```

```python
# Finding the number of null timestamps
null_cart_times = len(cart_visits[cart_visits.cart_time.isnull()])

print(null_cart_times)
```

```python
# From this information, it means that 1652 people out of the 2000 people who visited the site never made it to the cart 
null_cart_times
```

```python
# Finding the percentage of people who visited the cart 
percentage_no_cart_visit = null_cart_times/total_cart_visits 
percentage_no_cart_visit
```

```python
# The percentage of people who had a t-shirt placed in the cart but never made it to checkout
cart_checkout = cart.merge(checkout, how = "left")
cart_checkout
```

```python
# Checking out the null checkout times
null_times_of_checkout = len([cart_checkout.checkout_time.isnull()])
null_times_of_checkout
```

```python
cart_not_checkout = float(null_times_of_checkout) / float(len(cart))
cart_not_checkout
```

```python
# Percentage of users who got to checkout but did not purchase
reached_checkout = complete_data[~complete_data.checkout_time.isnull()]

checkout_not_purchase = complete_data[(purchases.purchase_time.isnull()) & (~complete_data.checkout_time.isnull())]

checkout_not_purchase_percent = float(len(checkout_not_purchase)) / float(len(reached_checkout))

print("% of users who got to checkout but did not purchase:",checkout_not_purchase_percent)
```
