```python
'''
The columns present in this dara are:
- `miles`: miles traveled through the flight
- `passengers`: number of passengers on the flight
- `delay`: take-off delay in minutes
- `inflight_meal`: is there a meal included in the flight?
- `inflight_entertainment`: are there free entertainment systems for each seat?
- `inflight_wifi`: is there complimentary wifi on the flight?
- `day_of_week`: day of the week of the flight
- `weekend`: did this flight take place on a weekend?
- `coach_price`: the average price paid for a coach ticket
- `firstclass_price`: the average price paid for first-class seats
- `hours`: how many hours the flight took
- `redeye`: was this flight a redeye (overnight)?
'''
```

```python
# import the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
# read in the data
flight_data = pd.read_csv("C:\\Users\\Lenovo\\OneDrive\\Documents\\UADSL\\PYTHON\DATA\\flight_data.csv")
flight_data
```

```python
# dataframe called coach_price
flight_data["coach_price"]
```

```python
## Task 2
print(np.mean(flight_data.delay[flight_data.delay == 8]))
print(np.median(flight_data.delay[flight_data.delay == 8]))

sns.histplot(flight_data.coach_price[flight_data.hours == 8])

plt.show()
plt.clf()
```

```python
# Visualizing the kinds of delays 
sns.histplot(flight_data.delay[flight_data.delay <= 500])
plt.show()
plt.clf()
```

```python
# visualizing the relationship between the coach prices and first-class proces.
perc = 0.1
flight_sub = flight_data.sample(n = int(flight_data.shape[0]*perc))

sns.lmplot(x = "coach_price", y = "firstclass_price", data = flight_sub, line_kws={'color': 'black'}, lowess=True)
plt.show()
plt.clf()
```

**MULTIVARIATE ANALYSIS**

```python
'''
The sns.Implot plots the linear regression models in seabbron. It helps visualize the relationship between two models by fitting a linear regression and displaying the scatter plots of the data
The syntax of sns.Implot is:
sns.Implot(x, y, data, hue = None, col = None, row = None, palette = None, markers = None, scatter = True, fit-reg = True, ci = .95, legend = True, height = 5, aspect = 1)
1. x = the name of the column of the dataframe to be plotted in the x-axis
2. y = the name of the column of the dataframe to be plotted in the y_axis
3. data = The dataframe the contains the data that you want to plot
4. hue = optional. grouping variable that will produce different regression lines and colors for the subsets of the data
5. col, row = optional. Variables that define the subset of the data plotted across different columns or rowns in a grid of plots
6. Markers : Optional: Marker style for scatter plots points
7. Scatter : Default is True. If set to False, the regression line will not be drawn
8. ci. size of the confidence interval for the regression line. The default is 0.95
9. legend : whether to add the legend. The default is True
10. Aspect : The aspect ration of each facet. The default s 1, so the plot is square 
```

```python
sns.lmplot(x = "coach_price", y = "firstclass_price", hue = "weekend", data = flight_sub, fit_reg = True, legend = True)
plt.show()
plt.clf()
```

```python
sns.lmplot(x = 'coach_price', y = 'firstclass_price', hue = 'weekend', data = flight_sub, fit_reg=False)
plt.show()
plt.clf()
```

```python
# How the coach prices change for the redeyes and the non-redeyes on each day of the week 
sns.boxplot(x = "day_of_week", y = "coach_price", hue = "redeye", data = flight_data)
plt.show()
plt.clf()
```
