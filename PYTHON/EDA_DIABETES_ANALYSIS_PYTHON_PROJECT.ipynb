### DATA CLEANING

```python
# checking if the data has any missing values
diabetis.describe()
```

```python
# checking the datatypes of the data
diabetis.info()
```

```python
# checking if any column has null values
diabetis.isnull().sum()
```

```python
# replace instances of 0 with NaN
diabetis[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = diabetis[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, np.NaN)
```

```python
# Checking the missing values that are present now
diabetis.isnull().sum()
```

```python
# print rows with missing values
print(diabetis[diabetis.isnull().any(axis=1)])
```

```python
diabetis[diabetis.isnull().any(axis=1)]
```

```python
# deletes all the missing values
diabetis = diabetis.dropna()
diabetis
```
